```python
def checkSequence(a,  b):
 
    # if length of the b = 0
    # then we return true
    if len(b) == 0:
        return True
 
    # if length of a = 0
    # that means b is not present in
    # a so we return false
    if len(a) == 0:
        return False
 
    if(a[0] == b[0]):
        return checkSequence(a[1:], b[1:])
    else:
        return checkSequence(a[1:], b)
 
 

s1 = "abchjsgsuohhdhyrikkknddg"
s2 = "coding"
 
if (checkSequence(s1, s2)):
    print("Yes")
else:
    print("No")
 
```
