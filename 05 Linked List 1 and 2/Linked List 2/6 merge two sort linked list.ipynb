```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
def printLL(head):
    while head is not None:
        print(str(head.data) + "->", end='')
        head=head.next
    print("None")
    return
def mergeSortedLists(head1,head2):
    
    FT=None
    
    if head1 is None:
        return head2
    if head2 is None:
        return head1

    if head1.data <= head2.data:
        FT=head1
        FT.next=mergeSortedLists(head1.next,head2)
    else:
        FT=head2
        FT.next=mergeSortedLists(head1,head2.next)
        

    return FT






    
def takeinput():
    inputlist=[int(ele) for ele in input().split()]
    head = None
    tail=None
    for currdata in inputlist:
        
        if currdata== -1:
            break
        newnode = Node(currdata)
        
        if head is None:
            head =newnode
            tail=newnode
        else:
            tail.next=newnode
            tail=newnode
      
    return head
head = takeinput()  
head2= takeinput()
printLL(head)
head=mergeSortedLists(head,head2)
printLL(head)
```
