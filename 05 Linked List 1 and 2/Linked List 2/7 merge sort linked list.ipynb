```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
def printLL(head):
    while head is not None:
        print(str(head.data) + "->", end='')
        head=head.next
    print("None")
    return

def mergeSortedLists(head1,head2):
    
    fh = None
    ft = None
    
    if head1 is None and head2 is None:
        return None
    if head1 is None:
        return head2
    if head2 is None:
        return head1
    
        
    if head1.data < head2.data :
        fh = head1
        ft = head1
        head1 = head1.next
    else:
        fh = head2
        ft = head2
        head2 = head2.next
    while (head1 is not None) and (head2 is not None):
        if (head1.data <= head2.data):
            ft.next = head1
            ft = ft.next
            head1 = head1.next
            
        else:
            ft.next = head2
            ft = ft.next
            head2 = head2.next
            
    while head1 is not None:
        
        ft.next = head1
        ft = ft.next
        head1 = head1.next
        
    while head2 is not None:
        
        ft.next = head2
        ft = ft.next
        head2 = head2.next
        
    return fh

def findmidpoint(head):

    if head is None or head.next is None:
        return head
    slow=head
    fast=head.next
    while fast is not None and fast.next is not None:
        slow=slow.next
        fast=fast.next.next
    print("mid point=",slow.data)
    return slow
    
def mergeSort(head):
    if head is None or head.next is None:
        return head
    
    mid=findmidpoint(head)
    starthead2 = mid.next
    mid.next=None
    
    lefthead1=mergeSort(head)
    righthead2=mergeSort(starthead2)
    
    Finaloutput=mergeSortedLists(lefthead1,righthead2)
    
    return Finaloutput
    
def takeinput():
    inputlist=[int(ele) for ele in input().split()]
    head = None
    tail=None
    for currdata in inputlist:
        
        if currdata== -1:
            break
        newnode = Node(currdata)
        
        if head is None:
            head =newnode
            tail=newnode
        else:
            tail.next=newnode
            tail=newnode
      
    return head
head = takeinput()  
# head2= takeinput()
printLL(head)
head=mergeSort(head)
printLL(head)
```
