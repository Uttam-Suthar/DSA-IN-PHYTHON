```python
class TreeNode:
    def __init__(self,data):
        self.data=data
        self.children=list()
        
        
def printTreeDetailed(root):
    # not a base case but an edge case
    if root==None:
        return
    print(root.data,":",end="")
    
    for child in root.children:
        print(child.data,",",end="")
        
    print()
    
    for child in root.children:
        printTreeDetailed(child)
        
import queue      
def takeTreeInputlevelwise():
    q = queue.Queue()
    print("Enter root")
    rootData=int(input())
    
    if rootData == -1 :
        return None
    root = TreeNode(rootData) 
    q.put(root)
    while (not q.empty()) :
        currenrNode = q.get()
        print("Enter num of children for",currenrNode.data)
        numchildren=int(input())
        for i in range(numchildren) :
            print("Enter next children for",currenrNode.data)
            childData=int(input())
            child = TreeNode(childData)
            currenrNode.children.append(child)
            q.put(child)
    
    return root
        
        
def takeTreeInput():
    print("Enter root data")
    rootData=int(input())
    if rootData==-1:
        return None
    
    root=TreeNode(rootData)
    print("Enter no of children for", rootData)
    childrenCount=int(input())
    for i in range(childrenCount):
        child=takeTreeInput()
        root.children.append(child)
    return root


def numNode(root):
    if root == None:
        return 0
    
    count=1
    
    for child in root.children:
        count=count+numNode(child)
    return count
```

```python
root=takeTreeInput()
printTreeDetailed(root)

print(numNode(root))
```
