```python
class BinarytreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
        
# def printbinaritree(root):
#     if root ==None:
#         return
#     print(root.data)
#     printbinaritree(root.left)
#     printbinaritree(root.right)
    
def printbinaritreedtails(root):
    if root ==None:
        return
    print(root.data,end=":")
    if root.left !=None:
        print("l",root.left.data,end=",")
    if root.right != None:
        print("R",root.right.data,end="")
    print()
    printbinaritreedtails(root.left)
    printbinaritreedtails(root.right)
    
    
# def greaterthanx(root,x):
#     if root==None:
#         return 0
#     leftcount=greaterthanx(root.left,x)
#     rightcount=greaterthanx(root.right,x)
#     count=leftcount+rightcount
    
#     if root.data>x:
#         count +=1
#     return count

def height(root):
    if root == None:
        return 0
    return 1+ max(height(root.left),height(root.right))

def diameter(root) :
    if root == None:
        return 0
    
    lh=height(root.left)
    rh=height(root.right)
    
    rootheight= 1+lh+rh
    
    leftD=diameter(root.left)
    rightD=diameter(root.right)
    
    rootDiameter=max(leftD,rightD)
    
    return max(rootheight,rootDiameter)

    
    
def treeinput():
    rootdata=int(input())
    if rootdata== -1:
        return None
    root=BinarytreeNode(rootdata)
    lefttree=treeinput()
    righttree=treeinput()
    root.left=lefttree
    root.right=righttree
    return root


def takeInput():
    print("Enter root data")
    rootData = int(input())
    if (rootData == -1):
        return None
    root = BinaryTreeNode(rootData)
    root.left = takeInput()
    root.right = takeInput()
    return root
```

```python
root=treeinput()
printbinaritreedtails(root)
print(diameter(root))
```
