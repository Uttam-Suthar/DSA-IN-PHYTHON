```python
MOD = 998244353

# Calculate the factorial of a number n modulo MOD
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact = (fact * i) % MOD
    return fact

# Calculate nCr (binomial coefficient) modulo MOD
def nCr(n, r):
    numerator = factorial(n)
    denominator = (factorial(r) * factorial(n - r)) % MOD
    denominator_inverse = pow(denominator, MOD - 2, MOD)  # Calculate modular inverse using Fermat's little theorem
    return (numerator * denominator_inverse) % MOD

# Calculate the number of valid ways to distribute politicians and economists
def calculate_ways(p, q, r):
    total_people = p + q
    total_rooms = r

    # Calculate the total number of ways without considering constraints
    total_ways = nCr(total_people, total_rooms)

    # Subtract the invalid distributions
    invalid1 = (total_rooms * nCr(p, 1) * nCr(q, 1)) % MOD
    invalid2 = (total_rooms * nCr(q, 1)) % MOD
    valid_ways = (total_ways - invalid1 - invalid2) % MOD

    return valid_ways

# Read the number of test cases
T = int(input())

# Process each test case
for _ in range(T):
    p, q, r = map(int, input().split())
    ways = calculate_ways(p, q, r)
    print(ways)

```

```python
def modulo_product(A, B):
    product = 1
    if A >= B:
        return 0
    else:
        for i in range(1, A + 1):
            product = (product * i) % B

    return product


T = int(input())
for _ in range(T):
    A, B = map(int, input().split())
    result = modulo_product(A, B)
    print(result)

Sample Input 1
4
8 10
5 140
18 19
20 21
Sample Output 1:
0
120
18
0
```
