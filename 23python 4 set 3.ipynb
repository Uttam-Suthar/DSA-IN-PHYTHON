```python
def countStrings(n, k):
    dp = [[[0, 0] for __ in range(k + 1)]
                  for _ in range(n + 1)]
 
    # If n = 1 and k = 0.
    dp[1][0][0] = 1
    dp[1][0][1] = 1
 
    for i in range(2, n + 1):
         
        # number of adjacent 1's can not exceed i-1
        for j in range(k + 1):
            dp[i][j][0] = (dp[i - 1][j][0] +
                           dp[i - 1][j][1])
            dp[i][j][1] = dp[i - 1][j][0]
            if j >= 1:
                dp[i][j][1] += dp[i - 1][j - 1][1]
 
    return dp[n][k][0] + dp[n][k][1]


T = int(input())

# Iterate over the test cases
for _ in range(T):
    # Read the input for each test case
    n, k = map(int, input().split())

    # Calculate and print the result
    result = countStrings(n, k)
    print(result)
```

```python
# write your code here
def solve_crossword(crossword, word_list):
    words = word_list.split(';')
    crossword = [list(row) for row in crossword]
    rows, cols = len(crossword), len(crossword[0])

    # Find the '-' cells in the crossword
    empty_cells = []
    for i in range(rows):
        for j in range(cols):
            if crossword[i][j] == '-':
                empty_cells.append((i, j))

    def is_valid(word, row, col, direction):
        if direction == 'across':
            if col + len(word) > cols:
                return False
            for k in range(len(word)):
                if crossword[row][col + k] != '-' and crossword[row][col + k] != word[k]:
                    return False
        elif direction == 'down':
            if row + len(word) > rows:
                return False
            for k in range(len(word)):
                if crossword[row + k][col] != '-' and crossword[row + k][col] != word[k]:
                    return False
        return True

    def place_word(word, row, col, direction):
        if direction == 'across':
            for k in range(len(word)):
                crossword[row][col + k] = word[k]
        elif direction == 'down':
            for k in range(len(word)):
                crossword[row + k][col] = word[k]

    def remove_word(word, row, col, direction):
        if direction == 'across':
            for k in range(len(word)):
                crossword[row][col + k] = '-'
        elif direction == 'down':
            for k in range(len(word)):
                crossword[row + k][col] = '-'

    def backtrack(idx):
        if idx == len(words):
            return True

        word = words[idx]
        for cell in empty_cells:
            row, col = cell

            if is_valid(word, row, col, 'across'):
                place_word(word, row, col, 'across')
                if backtrack(idx + 1):
                    return True
                remove_word(word, row, col, 'across')

            if is_valid(word, row, col, 'down'):
                place_word(word, row, col, 'down')
                if backtrack(idx + 1):
                    return True
                remove_word(word, row, col, 'down')

        return False

    # Start the backtracking algorithm
    if backtrack(0):
        return crossword

    return None

# Read the input
crossword = []
for _ in range(10):
    crossword.append(input().strip())

word_list = input().strip()

# Solve the crossword puzzle
solution = solve_crossword(crossword, word_list)

# Print the solution
if solution:
    for row in solution:
        print(''.join(row))
```

```python
# Power Pokemon Question

from collections import Counter

def solve(nums):
    MOD = int(1e9) + 7
    res, c = 0, Counter()
    for x in nums:
        for j in range(32):
            res = (res + c[(1 << j) - x]) % MOD
        c[x] += 1
    return res


n = int(input())
power = list(map(int, input().split()))

result = solve(power)
print(result)
```

```python
# Adjacent bit counts
MOD = 10**9 + 7

def countStrings(n, k):
    dp_prev = [[0, 0] for _ in range(k+1)]
    dp_prev[0][0] = 1
    dp_prev[0][1] = 1

    for i in range(2, n+1):
        dp_curr = [[0, 0] for _ in range(k+1)]
        for j in range(k+1):
            dp_curr[j][0] = (dp_prev[j][0] + dp_prev[j][1]) % MOD
            dp_curr[j][1] = (dp_prev[j][0] + dp_prev[j-1][1]) % MOD
        dp_prev = dp_curr

    return (dp_prev[k][0] + dp_prev[k][1]) % MOD


T = int(input())


for _ in range(T):
    n, k = map(int, input().split())

    result = countStrings(n, k)
    print(result)
```
